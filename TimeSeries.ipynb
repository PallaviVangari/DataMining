```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# Load your time series data into a DataFrame
# For this example, let's assume you have a CSV file with two columns: 'Date' and 'Sales'
data = pd.read_csv('raw_sales.csv')

# Convert the 'Date' column to a datetime format
data['datesold'] = pd.to_datetime(data['datesold'])

# Set 'Date' as the index
data.set_index('datesold', inplace=True)

# Plot using the 'Date' as the index
plt.figure(figsize=(12, 6))
plt.plot(data['price'])
plt.title('Sales Time Series')
plt.xlabel('Date')
plt.ylabel('price')
plt.show()
```
