```python

import numpy as np
import pandas as pd

dataset = pd.read_csv('/content/Global Missing Migrants Dataset.csv')
dataset.shape
dataset.head
```

```python
data = dataset.sample(frac=0.95, random_state=523)
data_unseen = dataset.drop(data.index)
data.reset_index(inplace=True, drop=True)
data_unseen.reset_index(inplace=True, drop=True)
print('Data for Modeling: ' + str(data.shape))
print('Unseen Data For Predictions: ' + str(data_unseen.shape))
```

```python
from pycaret.regression import *
regression = setup(data = data, target = 'Total Number of Dead and Missing', session_id=123,use_gpu=True)
```

```python
best = compare_models()
```
